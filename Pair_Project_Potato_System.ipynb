# Task 1 -  Number of Pokemon in each Generation 

```python
#Representing count(Pokemon) with respect to generation

alt.Chart(data2).mark_bar().encode(x = 'Generation:N',
                                           y = 'count_name:Q',
                                           color = 'Generation:N').transform_aggregate(count_name = 'count(Name):Q', 
                                        groupby=["Generation"]).properties(
                                                                title = 'Pokemon by Generation', 
                                                                width = 500, height = 500).configure_title(
                                                                fontSize = 20, font='Times New Roman').configure_legend(labelFontSize = 20, 
                                                                titleFontSize= 20).configure_axis(
                                                                labelFontSize = 15, titleFontSize = 15).interactive()
```

# Task 2- Type distribution For Each Generation

```python
brush = alt.selection_single(fields = ['Type 1'])

points = alt.Chart(data2).mark_bar().encode(
                x = alt.X('Generation:N'),
                y = alt.Y('count_Pokemon:Q'),
                color = alt.condition(brush, 'Type 1:N',
                                      alt.value('gray')),
                tooltip = ['Generation', 'Type 1', 'Name']
).add_selection(brush).transform_aggregate(
    count_Pokemon = 'count(Name):Q', groupby=['Generation', 'Type 1', 'Name']
).properties(width = 500, height = 500, title = 'Element Type 1 of Pokemon by Generation')


bars = alt.Chart(data2).mark_bar().encode(
y = 'count(Type 1):O',
x = 'Type 1:N',
color = alt.condition(brush,'Type 1:N',alt.value('gray')),
tooltip = ['count(Type 1)', 'Type 1']
).properties(title = 'Count of Selected Pokemon').transform_filter(brush).add_selection(brush)

points | bars
```

# Task 3 - Attack distribution by Type 1 and Legendary

```python
#min-max boxplot
alt.Chart(data2).mark_boxplot(extent='min-max').encode(
    x='Type 1:N',
    y ='Attack:Q',
    color = "Legendary:N").configure_legend(labelFontSize = 20, titleFontSize= 20).configure_axis(
                        labelFontSize = 15, titleFontSize = 15).properties(width = 500, height = 500,
                                     title = 'Attack Range of Pokemon by Type 1').configure_title(
                                     fontSize = 20, font = 'Times New Roman').interactive()
```

# Task 4 - Defense Distribution by Type 1 and Legendary

```python
#min-max boxplot
alt.Chart(data2).mark_boxplot(extent='min-max').encode(
    x='Type 1:N',
    y ='Defense:Q',
    color = "Legendary:N").properties(width = 500, height = 500,
                            title = "Defense range of Pokemon by type 1").configure_title(
                            fontSize = 20, font = 'Times New Roman').configure_legend(
                            labelFontSize = 20, titleFontSize= 20).configure_axis(
                            labelFontSize = 15, titleFontSize = 15).interactive()
```

# Task 5 - Top 10 non legendary pokemon with highest stats

```python
#Sorting by stats
data_pokemon = data2[data2['Legendary']== False]

#Extracting a subset of data and sorting
data_pokemon = data_pokemon.sort_values(by = ['Total'],ascending = False)
data_pokemon = data_pokemon[~data_pokemon['Name'].str.contains("Mega")].head(10)

#Renaming column names of Sp. Atk and Sp. Def
data_pokemon.rename(columns={'Sp. Atk':'Special_Attack', 'Sp. Def':'Special_Defense'}, inplace=True)
data_pokemon.head()
```

```python
#Heatmap to show the stats
alt.Chart(data_pokemon).mark_rect().encode(
        x = alt.X('Name:N', sort = '-y'),
        y = alt.Y('Total:Q'),
        tooltip=['Name','Attack','Defense','Special_Attack','Special_Defense', 'Type 1','Type 2'],
        color=alt.Color('Generation:N')
).properties(width = 400, 
             height = 400, title = 'Stats of top 10 Non Legendary Pokemon').configure_title(
             fontSize = 20, font = 'Times New Roman').configure_legend(labelFontSize = 20, 
                                                                       titleFontSize= 20).configure_axis(
                                                                       labelFontSize = 15, titleFontSize = 15)
```

```python
#Extracting Legendary pokemon Stats from the dataset
data_pokemon_legendary = data2[data2['Legendary']==True]
data_pokemon_legendary = data_pokemon_legendary[~data_pokemon_legendary['Name'].str.contains("Mega")]

#Renaming the columns
data_pokemon_legendary.rename(columns={'Sp. Atk':'Special_Attack', 'Sp. Def':'Special_Defense'}, inplace=True)
data_pokemon_legendary.head()
```

# Task 6- Legendary Pokemon stats by Generation

```python
alt.Chart(data_pokemon_legendary).mark_rect().encode(
        x=alt.X('Name:N', sort = '-y'),
        y='Total:Q',
        color='Generation:N',
        tooltip=['Name','Attack','Type 1','Type 2','Defense','Special_Attack','Special_Defense', 'Generation'],
).properties(title = 'Stats of Legendary Pokemon by Generation').configure_title(
                        fontSize = 20, font = 'Times New Roman').configure_legend(
                        labelFontSize = 20, titleFontSize= 20).configure_axis(
                        labelFontSize = 15, titleFontSize = 15)
```
